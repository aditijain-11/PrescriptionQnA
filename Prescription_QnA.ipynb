```python
import google.generativeai as genai
import PIL.Image
```

```python
genai.configure(api_key = Gemini_api_key)

model = genai.GenerativeModel('gemini-pro-vision', generation_config={
    "temperature":0.7,
    "top_p": 1,
    "top_k": 1,
    "max_output_tokens": 2048,
})
```

```python
img = PIL.Image.open('image.png')
img
```

```python
response = model.generate_content(["Provided the doctor prescription.what are multivitamins",img])
print(response.text)
```
